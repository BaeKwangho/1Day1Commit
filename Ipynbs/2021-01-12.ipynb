```python
import tensorflow as tf

space = len(word_json)+1
sos = len(word_json)+2
eos = len(word_json)+3
word_json['<sos>'] = sos
word_json['<eos>'] = eos
word_json[' '] = space

print(max_len)
sens = np.ndarray((len(text_list),max_len+2),dtype=int)
for i,k in enumerate(text_list):
    new_sen = [sos]+[word_json[word][0] for word in k]+[eos]
    for t in range(max_len - len(new_sen)):
        sens[i][:len(new_sen)+t] = space
    sens[i][:len(new_sen)] = new_sen
    
inputs = tf.cast(np.expand_dims(sens,axis=2),tf.float32)

print(sens.shape)
```

```python
lstm = tf.keras.layers.LSTM(512)
opt = tf.keras.optimizers.Adam(10e-2)

for sentence in inputs:
    with tf.GradientTape() as tape:
        result = lstm(tf.expand_dims(sentence,axis=1))
        loss = -tf.reduce_mean(tf.reduce_mean(result,axis=0))
    gradient = tape.gradient(loss,lstm.trainable_weights)
    opt.apply_gradients(zip(gradient,lstm.trainable_weights))
    print(loss)
```
